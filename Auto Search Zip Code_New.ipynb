```python
# Retrieve Zipcode

def google_geocode(address, api_key):
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        'address': address,
        'key': api_key
    }
    try:
        response = requests.get(url, params=params, timeout=10)
        data = response.json()
        if data['status'] == 'OK' and data['results']:
            result = data['results'][0]
            location = result.get('geometry', {}).get('location', {})
            lat = location.get('lat', None)
            lon = location.get('lng', None)
            zip_code = 'N/A'
            for comp in result.get('address_components', []):
                if 'postal_code' in comp['types']:
                    zip_code = comp['long_name']
            return lat, lon, zip_code
        else:
            print(f"  API error: {data['status']}")
            return None, None, 'N/A'
    except Exception as e:
        print(f"  Request failed: {e}")
        return None, None, 'N/A'
```

```python
import time
df['Latitude'] = ''
df['Longitude'] = ''
df['Zip Code'] = ''

for i, row in df.iterrows():
    address = f"{row['Account Name']}, {row['City']}, {row['State']}"
    lat, lon, zip_code = google_geocode(address, api_key)

    df.at[i, 'Latitude'] = lat
    df.at[i, 'Longitude'] = lon
    df.at[i, 'Zip Code'] = zip_code
    print(f"  Row {i+1}/{len(df)} processed: {zip_code}, {lat}, {lon}")
    time.sleep(1.2)  # Respectful pacing for API usage
```
