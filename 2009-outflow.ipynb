```python
import pandas as pd
df=pd.read_excel(r"C:\Users\EE\Downloads\2009 Shelter Stats_final.xlsx",  sheet_name="Outflow")
df.head(2)
```

```python
# Columns that apply to every row (not animal-specific)
fixed_cols = ['Pacfa ID', 'Name', 'Address', 'City', 'State', 'Zip']
```

```python
#Melt Dataframe
df_melted = df.melt(id_vars=fixed_cols, var_name='Animal Combo', value_name='Quantity')
df_melted
```

```python
#list of animals
animals = ['Dogs', 'Cats', 'Small Mammals', 'Reptiles', 'Pet Birds', 'Other']
```

```python
#Split column Animal Combo into Animal, Context
def split_animal_combo(combo):
    for animal in animals:
        if combo.startswith(animal):
            return animal, combo.replace(animal, "").strip()
    return None, combo  # fallback if no match
```

```python
# Apply split
df_melted[["Animal Name", "Context"]] = df_melted["Animal Combo"].apply(
    lambda x: pd.Series(split_animal_combo(x))
)
```
